```python
x = df[['YEAR']]
y = df[['JAN']]
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=1) 
print("x_train: ",x_train.head())
print("x_test: ",x_test.head())
print("y_train: ",y_train.head())
print("y_test: ",y_test.head())
```

```python
from sklearn import linear_model
ln = linear_model.LinearRegression()
ln.fit(x_train,y_train)
```

```python
y_pred_train = ln.predict(x_train)
y_pred_test = ln.predict(x_test)
```

```python
print("Intercept=" , ln.intercept_) 
print("Coefficient=" , ln.coef_)
```

```python
import numpy as np
MSE = np.square(np.subtract(y_test, y_pred_test)).mean() 
print("MSE= ",MSE)
MAE = np.abs(np.subtract(y_test, y_pred_test)).mean()
print("MAE= ", MAE)
RMSE = np.sqrt(MSE) 
print("RMSE= ", RMSE)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred_test)
print("mse= ",mse)
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y_test, y_pred_test)
print("rmse= ",rmse)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred_test)
print("mae= ",mae)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color="blue")
plt.plot(x_train, y_pred_train, color="red" , linewidth=5)
plt.title('Yearly Temperature')
plt.xlabel('YEAR')
plt.ylabel('TEMPERATURE(Celcius)')
plt.show()
```
